```python
# Checking Available data

from pycaret.datasets import get_data
from pycaret.datasets import *
import pandas as pd

dataSets = get_data('index')
```

```python
from pycaret.clustering import *

data = get_data('wholesale')
```

#  **K Means**

```python
# K means

# Setting up the environment
exp_clu = setup(data, normalize=True, session_id=123)

# List of cluster numbers to test
cluster_numbers = [3, 4, 5, 6]

# Loop over each cluster number
for num_clusters in cluster_numbers:
    print(f"\nNumber of Clusters: {num_clusters}")

    # Creating the KMeans clustering model with current number of clusters
    model = create_model('kmeans', num_clusters=num_clusters)

    # Plotting the elbow plot
    plot_model(model, plot='elbow')
    plot_model(model, plot='tsne')
```

# **Hierarchical Clustering**

```python
exp_clu = setup(data, normalize=True, session_id=123)

# List of cluster numbers to test
cluster_numbers = [3, 4, 5, 6]

# Loop over each cluster number
for num_clusters in cluster_numbers:
    print(f"\nNumber of Clusters: {num_clusters}")

    # Creating the Hierarchical Agglomerative Clustering model with current number of clusters
    model = create_model('hclust', num_clusters=num_clusters)

    # Plotting the elbow plot
    plot_model(model, plot='elbow')
    plot_model(model, plot='tsne')
```
